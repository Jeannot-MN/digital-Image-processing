```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from PIL import Image

img = np.array(Image.open('assets/images/profile.jpg').convert('L'))
plt.imshow(img, cmap=cm.gray)
plt.show()
```

```python
print(img)
print(img.shape)
```

```python
intensity_values = np.arange(0,256)
intensity_frequency = np.full(256, 0, 'uint')
for p in np.nditer(img):
    intensity_frequency[p] = intensity_frequency[p] + 1

np_hist, bins = np.histogram(img.flatten(), 256, (0, 255))
```

```python
# plt.hist(intensity_values,bins=500,data=intensity_frequency)
plt.bar(intensity_values, intensity_frequency)
plt.show()
```
