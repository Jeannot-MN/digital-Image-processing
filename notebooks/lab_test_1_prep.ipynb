# Convert Color to gray scale

```python
from skimage import color
```

```python
new_color_img = io.imread("lab1_imgs/lena_RGB.tif")
gray_from_color_img = color.rgb2gray(new_color_img)
io.imshow(gray_from_color_img)
gray_from_color_img.shape
```




# Affine transformation

```python
from scipy.ndimage import affine_transform
import numpy as np
from matplotlib.pyplot import cm
```

```python
# lena_gray = io.imread("lab1_imgs/lena_RGB.tif", as_gray=True) 
scaling_matrix = np.array([[.5, 0, 0], [0, 1, 0], [0, 0, 1]])
scaled_img = affine_transform(gray_from_color_img, scaling_matrix)

io.imshow(scaled_img, cmap='gray')
scaled_img.shape
```

```python
translation_matrix = np.array([[1, 0, 50], [0, 1, 50], [0, 0, 1]])
translated_img = affine_transform(gray_from_color_img, translation_matrix)

io.imshow(translated_img, cmap=cm.gray)
translated_img.shape
```

```python
translation_matrix = np.array([[-1, 0, 0], [0, -1, 0], [0, 0, 1]]) @ np.array([[1, 0, -512], [0, 1, -512], [0, 0, 1]])
translated_img = affine_transform(gray_from_color_img, translation_matrix)

io.imshow(translated_img, cmap=cm.gray)
translated_img.shape
```

```python
translation_matrix = np.array([[0.5,0,0],[0,1,0],[0,0,1]])

output_img = affine_transform(gray_from_color_img, translation_matrix)
io.imshow(output_img, cmap='gray')
```

```python
theta = -np.pi / 6

translation_matrix = np.array([[1, 0, 256], [0, 1,256], [0, 0, 1]]) @ np.array([[np.cos(theta), -np.sin(theta), 0], [np.sin(theta), np.cos(theta), 0], [0, 0, 1]]) @ np.array([[1, 0, -256], [0, 1, -256], [0, 0, 1]])
translated_img = affine_transform(gray_from_color_img, translation_matrix)

io.imshow(translated_img, cmap=cm.gray)
translated_img.shape
```
